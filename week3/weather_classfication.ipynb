```python
import pandas as pd

# -----------------------------
# 1) 데이터 불러오기
# -----------------------------
path = "/content/drive/MyDrive/ml-programming-lab/week3/weather.csv"
df = pd.read_csv(path)

print("원본 형태:", df.shape)
print("컬럼 목록:", list(df.columns))
```

```python
df
```

```python
# -----------------------------
# 정답 라벨 분포 확인
# -----------------------------
df['RainTomorrow'].value_counts()
```

```python
# -----------------------------
# 2) 결측치 확인
# -----------------------------
df.isnull().sum()

# -----------------------------
# 3) 결측치 제거
# -----------------------------
df = df.dropna()
df.isnull().sum()
```

```python
# -----------------------------
# 4) 모든 속성을 숫자로 변형
# -----------------------------


from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df
```

```python
# -----------------------------
# 5) X, y 분리
# -----------------------------
X = df.drop(columns=["RainTomorrow"])
y = df["RainTomorrow"]
```

```python
X
```

```python
y
```

```python
# -----------------------------
# 6) 학습/평가 세트 분리
# -----------------------------
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
df
```

```python
# -----------------------------
# 7) 모델 구성
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=500)
```

```python
# -----------------------------
# 8) 모델 학습
# -----------------------------
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
```

```python
# -----------------------------
# 9) 모델 예측
# -----------------------------
dt_y_pred = dt.predict(X_test)
rf_y_pred = rf.predict(X_test)
lr_y_pred = lr.predict(X_test)

# -----------------------------
# 정답/예측값 일부 확인
# -----------------------------
print("\n=== y_test (head) ===")
print(y_test.head().values)

print("\n=== DT 예측 (head) ===")
print(dt_y_pred[:10])

print("\n=== RF 예측 (head) ===")
print(rf_y_pred[:10])

print("\n=== LR 예측 (head) ===")
print(lr_y_pred[:10])
```

```python
# -----------------------------
# 10) 모델 평가
# -----------------------------
from sklearn.metrics import accuracy_score, confusion_matrix

dt_acc = accuracy_score(y_test, dt_y_pred)
rf_acc = accuracy_score(y_test, rf_y_pred)
lr_acc = accuracy_score(y_test, lr_y_pred)

print("\n=== DT Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print(confusion_matrix(y_test, dt_y_pred))

print("\n=== RF Test Accuracy ===")
print(f"Random Forest : {rf_acc:.4f}")
print(confusion_matrix(y_test, rf_y_pred))

print("\n=== LR Test Accuracy ===")
print(f"Logistic Reg. : {lr_acc:.4f}")
print(confusion_matrix(y_test, lr_y_pred))
```
