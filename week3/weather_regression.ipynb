```python
import pandas as pd

# -----------------------------
# 1) 데이터 불러오기
# -----------------------------
path = "/content/drive/MyDrive/ml-programming-lab/week3/weather.csv"
df = pd.read_csv(path)

print("원본 형태:", df.shape)
print("컬럼 목록:", list(df.columns))
```

```python
df
```

```python
# -----------------------------
# 2) 열 삭제 (RainTomorrow 제거)
# -----------------------------
if "RainTomorrow" in df.columns:
    df = df.drop(columns=["RainTomorrow"])
```

```python
df
```

```python
# -----------------------------
# 정답 라벨 분포 확인
# -----------------------------
df['Rainfall'].value_counts()
```

```python
# -----------------------------
# 3) 결측치 확인
# -----------------------------
df.isnull().sum()

# -----------------------------
# 4) 결측치 제거
# -----------------------------
df = df.dropna()
df.isnull().sum()
```

```python
# -----------------------------
# 5) 모든 속성을 숫자로 변형
# -----------------------------


from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df
```

```python
# -----------------------------
# 6) X, y 분리
# -----------------------------
X = df.drop(columns=["Rainfall"])
y = df["Rainfall"]
```

```python
X
```

```python
y
```

```python
# -----------------------------
# 7) 학습/평가 세트 분리
# -----------------------------

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# -----------------------------
# 8) 모델 구성
# -----------------------------

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

lr = LinearRegression()
rf = RandomForestRegressor()
dt = DecisionTreeRegressor()
```

```python
# -----------------------------
# 9) 모델 학습
# -----------------------------

lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
dt.fit(X_train, y_train)
```

```python
# -----------------------------
# 10) 모델 예측
# -----------------------------
dt_y_pred = dt.predict(X_test)
rf_y_pred = rf.predict(X_test)
lr_y_pred = lr.predict(X_test)

# -----------------------------
# 정답/예측값 일부 확인
# -----------------------------
print("\n=== y_test (head) ===")
print(y_test.head().values)

print("\n=== DT 예측 (head) ===")
print(dt_y_pred[:10])

print("\n=== RF 예측 (head) ===")
print(rf_y_pred[:10])

print("\n=== LR 예측 (head) ===")
print(lr_y_pred[:10])
```

```python
# -----------------------------
# 11) 모델 평가
# -----------------------------
from sklearn.metrics import mean_squared_error, r2_score

# Decision Tree
dt_mse = mean_squared_error(y_test, dt_y_pred)
dt_r2 = r2_score(y_test, dt_y_pred)

# Random Forest
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r2 = r2_score(y_test, rf_y_pred)

# Linear Regression
lr_mse = mean_squared_error(y_test, lr_y_pred)
lr_r2 = r2_score(y_test, lr_y_pred)

print("\n=== 평가 결과 ===")
print(f"Decision Tree     -> MSE: {dt_mse:.4f}, R²: {dt_r2:.4f}")
print(f"Random Forest     -> MSE: {rf_mse:.4f}, R²: {rf_r2:.4f}")
print(f"Linear Regression -> MSE: {lr_mse:.4f}, R²: {lr_r2:.4f}")
```
