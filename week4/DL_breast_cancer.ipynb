```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# -----------------------------
# 1) 데이터 불러오기
# -----------------------------

df = pd.read_csv("/content/drive/MyDrive/ml-programming-lab/week4/breast_cancer.csv", index_col=0)
```

```python
df
```

```python
# 레이블 갯수 확인
df['label'].value_counts()
```

```python
# -----------------------------
# 2) 결측치 확인
# -----------------------------

df.isnull().sum()

# -----------------------------
# 3) 결측치 제거
# -----------------------------

df = df.dropna()
df.isnull().sum()
```

```python
# -----------------------------
# 4) 특징(X)과 타겟(y) 분리
# -----------------------------

X=df.drop('label',axis=1)
y=df['label']
```

```python
X
```

```python
y
```

```python
# -----------------------------
# 5) 타겟 변수 원-핫 인코딩
# -----------------------------

Y = pd.get_dummies(y).values
```

```python
# -----------------------------
# 6) 딥러닝 입력을 위해 Numpy로 변환
# -----------------------------

X = X.values
```

```python
X
```

```python
Y
```

```python
# -----------------------------
# 7) 훈련 데이터와 테스트 데이터 분리
# -----------------------------

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# -----------------------------
# 8) 딥러닝 모델 구성 (Dropout 레이어 제외)
# -----------------------------

model = Sequential()

model.add(Dense(10,input_shape=(30,),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
# -----------------------------
# 9) 모델 컴파일
# -----------------------------

model.compile(
    optimizer=Adam(learning_rate=0.04),
    loss='categorical_crossentropy',  # 원-핫 인코딩된 다중 분류 문제에 사용
    metrics=['accuracy'])

# 모델 요약 정보 출력
model.summary()
```

```python
# -----------------------------
# 10) 모델 학습
# -----------------------------

model_history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
# -----------------------------
# 11) 모델 예측
# -----------------------------
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
# -----------------------------
# 12) 모델 평가
# -----------------------------

print("\n< Classification Report >")
print(classification_report(y_test_class, y_pred_class, digits=4))

print("< Confusion Matrix (rows=true, cols=pred) >")
print(confusion_matrix(y_test_class, y_pred_class))
```
